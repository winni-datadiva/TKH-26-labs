## âœ… Setup: Create a "population"
We will create a population of 5,000 values representing spend amounts.

```python
import random
import statistics as stats

random.seed(7)

# Population: spend amounts (mix of lower + higher spenders)
N = 5000
population = []
for _ in range(N):
    if random.random() < 0.2:   # 20% higher spenders
        spend = random.gauss(80, 20)
    else:                      # 80% typical spenders
        spend = random.gauss(35, 10)
    population.append(max(0, spend))

print("Population size:", len(population))
print("First 5 values:", population[:5])
```
