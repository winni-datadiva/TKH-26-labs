## 3) Data types
Python has different types for different kinds of data.

- `str` (string): text
- `int` (integer): whole numbers
- `float`: decimals
- `bool`: `True` or `False`

```python
city = "New York"
temperature = 72
gpa = 3.75
is_student = True

print(type(city))
print(type(temperature))
print(type(gpa))
print(type(is_student))
```

## 4) Basic math
Python can do arithmetic like a calculator.

Operators:
- `+` add
- `-` subtract
- `*` multiply
- `/` divide
- `**` exponent (power)
- `%` remainder (modulo)

```python
x = 10
y = 3

print("x + y =", x + y)
print("x - y =", x - y)
print("x * y =", x * y)
print("x / y =", x / y)
print("x ** y =", x ** y)
print("x % y =", x % y)
```
