## Step 5: Fix the Distance Column

Goal:
- Convert the MRT distance column into a numeric type so we can analyze it.

Recall that when we tried to convert this with pd.to_numeric() we discovered some hidden chracters, and so we needed to find a different way to convert this column to a numeric data type.

- We first ensure that the series is of type string by typecasting it via astype(str)
- we then access pandas string replace method with str.replace()
- next we replace the problematic characters
- lastly we convert the series to type float with .astype(float)

```python
df['X3 distance to the nearest MRT station'] = df['X3 distance to the nearest MRT station'].astype(str).str.replace('"', '').astype(float)
```
